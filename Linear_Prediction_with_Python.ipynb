```
import numpy as np
import matplotlib.pyplot as plt
```

```
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]

#our model for the forward pass
def forward(x):
  return x * w
```

```
#loss function
def loss(x, y):
  y_pred = forward(x)
  return (y_pred - y) * (y_pred -y)
```

```
#Defining empty Loss and Mean Square Error lists for the model.
w_list = []
mse_list = []
print(np.arange(0.0,4.1,0.1))
```

```
for w in np.arange(0.0,4.1,0.1):
  print("w = ", w)
  l_sum = 0
  #Zip input and output and then read the values from the zipped list
  for x_val,y_val in zip(x_data,y_data):
    y_pred_val = forward(x_val)
    l = loss(y_pred_val,y_val)
    l_sum += l
    print('x_in = {}, y_out = {}, y_pred = {}, loss = {}'.format(x_val,y_val,y_pred_val,l))
    #Divide the loss by total number of instances.
  print("MSE = ", l_sum / 3)
  w_list.append(w)
  mse_list.append(l_sum / 3)
  
```

```
plt.plot(w_list,mse_list)
plt.ylabel('Loss')
plt.xlabel('Weight')
plt.show()
```
